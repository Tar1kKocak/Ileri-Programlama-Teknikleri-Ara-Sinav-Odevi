```python
import random
import matplotlib.pyplot as plt
import numpy as np

x = 0
y = 14                  # Okul numaramın son iki hanesi 14 - 22040201014

topilkxkoordinat = [x]
topilkykoordinat = [y]
topsonxkoordinat = [x]
topsonykoordinat = [y]

g = 10                  # Yerçekimi
atisacisi = 30
min_v = 330             #Minimum Hız
max_V = 1800            #Maksimum Hız

haritamesafe = 20000 + 200 * (random.randint(-10, 10))
haritagenislikbaslangic = haritamesafe
haritagenislikbitis = haritamesafe + 1000 + 100 * (random.randint(-2, 2))

haritaxkoordinat = [haritagenislikbaslangic, haritagenislikbitis]
haritaykoordinat = [0, 0]

dongudurdurucu = 1
atisdenemesi = 1

while dongudurdurucu == 1:

    baslangic_V = (min_v + max_V) / 2

    T_Ucus = int((2 * baslangic_V * np.sin(np.radians(atisacisi))) / g) + 2

    formulegoresonx = round(x + baslangic_V * np.cos(np.radians(atisacisi)) * (2 * baslangic_V * np.sin(np.radians(atisacisi))/g))

    for a in range(0, T_Ucus + 1):      # Yanlış Atışlar Grafiği İçin Grafik Çizimi

        topharitasonux = round(x + baslangic_V * np.cos(np.radians(atisacisi)) * a)
        topharitasonuy = round(y + (baslangic_V * np.sin(np.radians(atisacisi)) * a - 0.5 * g * a ** 2))
        topilkxkoordinat.append(topharitasonux)
        topilkykoordinat.append(topharitasonuy)

    if formulegoresonx > haritagenislikbaslangic and formulegoresonx < haritagenislikbitis:

        print(atisdenemesi, ". seferde vurus gerceklesmistir . hedefi vurmak icin gerekli hiz ", baslangic_V)
        son_hiz = baslangic_V
        dongudurdurucu = 0

        for b in range(0, T_Ucus):      # Doğru Atışlar Grafiği İçin Grafik Çizimi

            topharitasonux = round(x + baslangic_V * np.cos(np.radians(atisacisi)) * b)
            topharitasonuy = round(y + (baslangic_V * np.sin(np.radians(atisacisi)) * b - 0.5 * g * b ** 2))
            topsonxkoordinat.append(topharitasonux)
            topsonykoordinat.append(topharitasonuy)

        #  1.TABLO    (DOĞRU ATIŞLAR)
        plt.subplot(1, 2, 1)
        plt.xlim(-1500, 30000)
        plt.ylim(-500, 5000)
        plt.title("Doğru Atış Grafiği")
        plt.xlabel("Mesafe (m)")
        plt.ylabel("Yükseklik (m)")
        plt.plot(topsonxkoordinat, topsonykoordinat,marker='*', color='cyan')
        plt.plot(haritaxkoordinat, haritaykoordinat, color='black')
        plt.plot(x, y, marker='*', color='black')

        #  2.TABLO    (YANLIŞ ATIŞLAR)
        plt.subplot(1, 2, 2)
        plt.title("Yapılan Atışlar")
        #plt.xlim(-1500, 30000)
        #plt.ylim(-500, 100000)
        plt.xlabel("Mesafe (m)")
        #plt.ylabel("Yükseklik (m)")
        plt.plot(topsonxkoordinat, topsonykoordinat, marker='*', color='cyan')
        plt.plot(topsonxkoordinat, topsonykoordinat, marker='*', color='cyan')
        plt.plot(topsonxkoordinat, topsonykoordinat, marker='*', color='cyan')
        plt.plot(topilkxkoordinat, topilkykoordinat, color='green')
        plt.plot(x, y, marker='*', color='black')

        plt.show()
        break

    elif formulegoresonx < haritagenislikbaslangic:

        print("onune dustu")
        min_v = int(baslangic_V)
        atisdenemesi = atisdenemesi + 1

    elif formulegoresonx > haritagenislikbitis:
        print("uzagina dustu")
        max_V = int(baslangic_V)
        atisdenemesi = atisdenemesi + 1
```
